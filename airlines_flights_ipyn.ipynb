# Task
Visualize the `time_bins_by_city` DataFrame as a heatmap to show flight counts based on source city, destination city, departure time, and arrival time.

## Prepare the data for visualization

### Subtask:
Reshape the `time_bins_by_city` DataFrame to be suitable for a heatmap, potentially aggregating further if the number of unique combinations is too large for a clear visualization.


**Reasoning**:
Pivot the `time_bins_by_city` DataFrame to prepare it for heatmap visualization.



```python
heatmap_data = time_bins_by_city.pivot_table(
    index=['source_city', 'destination_city'],
    columns=['departure_time', 'arrival_time'],
    values='count',
    fill_value=0
)
display(heatmap_data.head())
```

## Visualize the data

### Subtask:
Create a heatmap to show the flight counts based on source city, destination city, departure time, and arrival time using the prepared `heatmap_data`.


**Reasoning**:
Create a heatmap to visualize the flight counts based on source city, destination city, departure time, and arrival time using the prepared heatmap_data.



```python
plt.figure(figsize=(20, 15))
sn.heatmap(heatmap_data, annot=False, cmap='viridis')
plt.title('Flight Counts by City Pair, Departure Time, and Arrival Time')
plt.tight_layout()
plt.show()
```

## Refine the visualization

### Subtask:
Adjust the heatmap settings for better readability if needed.


**Reasoning**:
The current heatmap is too large to display annotations effectively. To improve readability, I will remove the annotations and slightly adjust the figure size.



```python
plt.figure(figsize=(25, 18))
sn.heatmap(heatmap_data, annot=False, cmap='viridis')
plt.title('Flight Counts by City Pair, Departure Time, and Arrival Time')
plt.tight_layout()
plt.show()
```

## Summary:

### Data Analysis Key Findings
