## Data visualization

### Subtask:
Visualize the distribution of features and the relationship between relevant columns.


**Reasoning**:
Visualize the distribution of selected numerical features and the target variable, and explore the relationships between relevant numerical features using scatter plots and a correlation heatmap.



```python
# Step 1: Visualize distribution of selected numerical features
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
df['radius_mean'].hist(bins=20)
plt.title('Distribution of Radius Mean')

plt.subplot(1, 3, 2)
sns.boxplot(x=df['texture_mean'])
plt.title('Distribution of Texture Mean')

plt.subplot(1, 3, 3)
df['perimeter_mean'].hist(bins=20)
plt.title('Distribution of Perimeter Mean')
plt.tight_layout()
plt.show()

# Step 2: Visualize distribution of the target variable
plt.figure(figsize=(6, 4))
sns.countplot(x=df['diagnosis'])
plt.title('Distribution of Diagnosis')
plt.show()

# Step 3: Visualize the relationship between a few pairs of relevant numerical columns
plt.figure(figsize=(10, 6))
sns.scatterplot(x='radius_mean', y='area_mean', hue='diagnosis', data=df)
plt.title('Radius Mean vs Area Mean by Diagnosis')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='perimeter_mean', y='texture_mean', hue='diagnosis', data=df)
plt.title('Perimeter Mean vs Texture Mean by Diagnosis')
plt.show()

# Step 4: Visualize the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=False, cmap='coolwarm')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
