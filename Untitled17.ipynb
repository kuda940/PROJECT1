```python
# visualize sales by years
```

```python
walmart_sales.groupby('Year')['Weekly_Sales'].sum().plot(kind='bar')
plt.title('Sales by Year')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

```python
# Get the top 5 stores based on total weekly sales
top_5_stores = walmart_sales.groupby('Store_Number')['Weekly_Sales'].sum().nlargest(5).index

# Filter the dataframe to include only the top 5 stores
top_5_sales_data = walmart_sales[walmart_sales['Store_Number'].isin(top_5_stores)]

# Group by store and year and sum the weekly sales
sales_by_store_year = top_5_sales_data.groupby(['Store_Number', 'Year'])['Weekly_Sales'].sum().unstack()

# Plot the data
sales_by_store_year.plot(kind='bar', figsize=(12, 6))
plt.title('Weekly Sales of Top 5 Stores by Year')
plt.xlabel('Store Number')
plt.ylabel('Total Weekly Sales')
plt.xticks(rotation=0)
plt.legend(title='Year')
plt.tight_layout()
plt.show()
```
