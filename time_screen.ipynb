```python
bins = [8, 11, 15, 18]
labels = ['Preteen', 'Early Teen', 'Teenager']
df['Age_Bucket'] = pd.cut(df['Age'], bins=bins, labels=labels, right=True, include_lowest=True)

display(df.head())
```

```python
df['Age_Bucket'].value_counts()
```

```python
df.groupby('Age_Bucket')['Avg_Daily_Screen_Time_hr'].mean()
```

```python
#visualize with a pie chart the above information
```

```python
age_bucket_counts = df['Age_Bucket'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(age_bucket_counts, labels=age_bucket_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Age Buckets')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
avg_screen_time_by_age_bucket = df.groupby('Age_Bucket', observed=False)['Avg_Daily_Screen_Time_hr'].mean()

plt.figure(figsize=(8, 6))
avg_screen_time_by_age_bucket.plot(kind='bar')
plt.title('Average Daily Screen Time by Age Bucket')
plt.xlabel('Age Bucket')
plt.ylabel('Average Daily Screen Time (hr)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
health_impacts_by_age_bucket = df.groupby('Age_Bucket', observed=False)['Health_Impacts'].value_counts().unstack().fillna(0)
display(health_impacts_by_age_bucket)

health_impacts_by_age_bucket.plot(kind='bar', stacked=True, figsize=(12, 7))
plt.title('Distribution of Health Impacts by Age Bucket')
plt.xlabel('Age Bucket')
plt.ylabel('Number of Children')
plt.xticks(rotation=0)
plt.legend(title='Health Impacts', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
health_impacts_by_device = df.groupby('Primary_Device')['Health_Impacts'].value_counts().unstack().fillna(0)
display(health_impacts_by_device)

health_impacts_by_device.plot(kind='bar', stacked=True, figsize=(12, 7))
plt.title('Distribution of Health Impacts by Primary Device')
plt.xlabel('Primary Device')
plt.ylabel('Number of Children')
plt.xticks(rotation=0)
plt.legend(title='Health Impacts', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
health_impacts_by_urban_rural = df.groupby('Urban_or_Rural')['Health_Impacts'].value_counts().unstack().fillna(0)
display(health_impacts_by_urban_rural)

health_impacts_by_urban_rural.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Distribution of Health Impacts by Urban or Rural Area')
plt.xlabel('Urban or Rural')
plt.ylabel('Number of Children')
plt.xticks(rotation=0)
plt.legend(title='Health Impacts', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
